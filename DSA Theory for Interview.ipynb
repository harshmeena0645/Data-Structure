```python
ARRAY:-
    "An array is a data structure. It is a collection of similar items stored in contiguous memory locations."
    üß† Property	 Fixed Size, Same Data Type, Indexed Access, Contiguous Memory
    üìö Type	 1D Array, 2D Array, Multidimensional, Static Array, Dynamic Array
    üõ†Ô∏è Operation	 Access, Update, Insert (end,middle), Delete (end,middle), Search (linear,binary), Sort
    
    ‚úÖ Advantages: 
                    Easy to implement
                    Fast random access
                    Efficient memory usage for static-size collections

    ‚ùå Disadvantages: 
                    Fixed size in static arrays
                    Insertion/deletion in the middle is expensive (O(n))
    
    üåü Applications:
                    Implementing Other Data Structures
                    Sorting and Searching Algorithms
                    GPS Navigation Systems
                    Image Processing
    üß† When Use Arrays?
                    You need indexed access
                    You want to store elements of same type
                    You don‚Äôt frequently insert/delete from the middle
                    You need efficient memory layout
```

```python
STACK:-
    "A Stack is a linear data structure that follows the LIFO (Last In, First Out) principle."
    üß† Properties:
                Order	Last In, First Out (LIFO)
                Insertion	Done at the top
                Deletion	Done from the top
                Linear Data Structure
                Same Data Type
    üìö Types : Simple Stack, Dynamic Stack, Stack of Stacks, Two Stacks in One Array
    üõ†Ô∏è Operations: push(x), pop(), peek()/top(), isEmpty(), isFull()
    ‚öôÔ∏è Implementation: Using Array, Using LL
        
    ‚úÖ Advantages:
                Easy to implement
                Fast operations (O(1) time for push/pop)
                Helps in recursion and backtracking problems

    ‚ùå Disadvantages:
                Fixed size in array-based stack
                Cannot access middle elements directly
    üåü Applications:
                Undo/Redo in Editors
                Backtracking Algorithms
```

```python
QUEUE:-
    "A Queue is a linear data structure that follows the FIFO (First In, First Out) principle."
    üß† Properties:
                Order	First In, First Out (FIFO)
                Insertion (Enqueue)	At the rear (or end)
                Deletion (Dequeue)	From the front
                Linear Data Structure
                Same Data Type
    üìö Types : Simple Queue, Circular Queue, Priority Queue, Double-Ended Queue (Deque)
    üõ†Ô∏è Operations: enqueue(item), dequeue(), peek()/front(), isEmpty(), isFull()
    ‚öôÔ∏è Implementation: Using Array, Using LL
        
    ‚úÖ Advantages:
                Easy to implement
                Useful in scheduling and real-world modeling

    ‚ùå Disadvantages:
                Limited in flexibility if using arrays
                Inefficient for middle operations
                
    üåü Applications:
                Handling Requests in Web Servers
                Ticket Booking Systems
                CPU and Disk Scheduling
```

```python
LINKED LIST:-
    "A Linked List is linear data structure where elements are stored in memory non-contiguously and connected using pointers."
    üß† Properties:
                Each node has data and a pointer
                Non-contiguous memory allocation
                Dynamic size
                Sequential access only
                Efficient insertions/deletions
    üìö Types : Singly Linked List, Doubly Linked List, Circular Linked List, Circular Doubly Linked List
    üõ†Ô∏è Operations: insert at start(), insert at end(), insert at position(p)
                    delete from start(), delete from end(), delete at position(p)
                    search(value), display()
    ‚úÖ Advantages:
                Dynamic size
                Efficient insertion/deletion
                No memory wastage (non-contiguous memory usage)

    ‚ùå Disadvantages:
                No random access (must traverse to find elements)
                More memory (pointers need extra space)
                Extra complexity in implementation
        
    üåü Applications:
                Music Player Playlists
                Implementing Stacks and Queues
                Image Viewer
                Job Scheduling in OS
                Undo Functionality
```

```python
TREE:-
    "A Tree is a non-linear, hierarchical data structure that consists of nodes connected by edges, 
    with the following key properties:
    One node is designated as the root.
    Each node (except the root) has exactly one parent.
    A node can have zero or more children.
    There are no cycles (it is acyclic).
    It is a connected graph with N nodes and N-1 edges.
                         
    üß± Basic Terminology                     
    | Term        | Explanation                                     |
    | ----------- | ----------------------------------------------- |
    | **Node**    | A single element of the tree (contains data)    |
    | **Root**    | The topmost node of the tree                    |
    | **Child**   | A node that descends from another node          |
    | **Parent**  | A node that has children                        |
    | **Leaf**    | A node with no children                         |
    | **Edge**    | A link between parent and child                 |
    | **Subtree** | Any child node along with its descendants       |
    | **Depth**   | Number of edges from the root to a node         |
    | **Height**  | Number of edges from a node to the deepest leaf |
    | **Level**   | Distance from the root (root is level 0)        |
    | **Degree**  | Number of children a node has                   |

    üìö Types of Trees
    1. Binary Tree: Each node has at most 2 children (left and right)
    2. Binary Search Tree (BST): Left child < Parent < Right child
    3. Balanced Tree: Height is kept minimal for faster operations (e.g., AVL Tree, Red-Black Tree)
    4. Heap: A complete binary tree:
            Max Heap: Parent ‚â• Children
            Min Heap: Parent ‚â§ Children
    5. Trie (Prefix Tree): Used for storing strings (like dictionaries, autocomplete)
    6. N-ary Tree: Each node can have up to N children.

    üîÅ Tree Traversals
    Ways to visit all nodes in a tree:
        ‚û§ Depth First Search (DFS):
        Inorder (L, Root, R): Left ‚Üí Root ‚Üí Right
        Preorder (Root, L, R): Root ‚Üí Left ‚Üí Right
        Postorder (L, R, Root): Left ‚Üí Right ‚Üí Root

        ‚û§ Breadth First Search (BFS):
        Level Order Traversal (uses Queue)
    
    ‚úÖ Advantages:
            Represents hierarchical relationships clearly (like file systems or organization charts).
            Efficient search, insert, and delete operations in balanced trees (O(log n)).
            In-order traversal of Binary Search Trees gives sorted data.

    ‚ùå Disadvantages:
            Implementation is more complex compared to arrays or linked lists.
            Performance degrades to O(n) if the tree is unbalanced.
            Requires extra memory for storing child pointers.

    üåü Applications: 
            File systems and folders
            HTML/XML DOM structure
            Binary Search Trees for fast search
            Expression trees in compilers
            Tries for autocomplete/spell-check
            AI and game theory (decision trees)
            Databases (e.g., B-trees, B+ trees)

    üìà Time Complexity (for Binary Trees)                         
        | Operation | Average Case | Worst Case |
        | --------- | ------------ | ---------- |
        | Search    | O(log n)     | O(n)       |
        | Insert    | O(log n)     | O(n)       |
        | Delete    | O(log n)     | O(n)       |
        For balanced trees like AVL or Red-Black Trees, time stays close to O(log n).                         
```

```python
GRAPH:-
    A Graph is a non-linear data structure made up of:
        Vertices (or Nodes) ‚Äì The individual entities.
        Edges ‚Äì The connections or relationships between those entities.
    Graphs are used to model networks: social networks, transportation systems, maps, etc.
    
    üß± Graph Terminology
    | Term                  | Meaning                                                         |
    | --------------------- | --------------------------------------------------------------- |
    | **Vertex**            | A point or node in the graph.                                   |
    | **Edge**              | A line connecting two vertices (can be directed or undirected). |
    | **Adjacent Vertices** | Two vertices connected by an edge.                              |
    | **Degree**            | Number of edges connected to a vertex.                          |
    | **Path**              | A sequence of vertices connected by edges.                      |
    | **Cycle**             | A path that starts and ends at the same vertex.                 |
    | **Connected Graph**   | There‚Äôs a path between every pair of vertices.                  |

    üìö Types of Graphs
    1. Directed Graph (Digraph)
        Edges have a direction (‚Üí)
        Example: Twitter following
    2. Undirected Graph
        Edges don‚Äôt have a direction
        Example: Facebook friends
    3. Weighted Graph
        Each edge has a weight/cost
        Example: Distance between cities
    4. Unweighted Graph
        All edges are treated equally
    5. Cyclic vs Acyclic
        Cyclic: Contains at least one cycle
        Acyclic: No cycles
    6. Connected vs Disconnected
        Connected: All nodes are reachable
        Disconnected: Some nodes can not be reached from others

    üß≠ Graph Representations
        Adjacency Matrix: A 2D array where matrix[i][j] = 1 if there is an edge from vertex i to j.
        Adjacency List: An array/list of lists. Each index holds a list of connected vertices.

    üîÅ Graph Traversal
        Depth First Search (DFS): Uses a stack (or recursion). Goes deep into the graph.
        Breadth First Search (BFS): Uses a queue. Explores neighbors level by level.
        
    ‚úÖ Advantages:
            Perfect for representing real-world relationships.
            Flexible with both direction and weight support.
            Efficient algorithms for traversal and pathfinding (like BFS, DFS, Dijkstra).

    ‚ùå Disadvantages:
            More complex to implement and manage than arrays or trees.
            Requires more memory (especially adjacency matrix for sparse graphs).
            Harder to debug due to cycles and complex connections.
        
    üí° Applications:
            Social networks (Facebook, LinkedIn)
            Web page ranking (Google uses graphs)
            Maps and GPS (shortest path algorithms like Dijkstra)
            Network routing
            Course prerequisite dependencies
            AI (state-space search)
        
    ‚öôÔ∏è Time Complexity:
        | Operation          | Adjacency Matrix | Adjacency List |
        | ------------------ | ---------------- | -------------- |
        | Space              | O(V¬≤)            | O(V + E)       |
        | Check edge (u, v)  | O(1)             | O(V)           |
        | Traverse all edges | O(V¬≤)            | O(V + E)       |
```

```python
HTML(HyperText Markup Language):-
    "It is the standard language used to create and design the structure of web pages."
    "It tells the browser how to display text, images, videos, links, forms, and more."
        | Concept        | Description                                                            |
        | -------------- | ---------------------------------------------------------------------- |
        | **HyperText**  | Links between web pages                                                |
        | **Markup**     | Tags used to define elements (e.g., <p> , <h1>)                        |
        | **Elements**   | Building blocks of an HTML document                                    |
        | **Tags**       | Keywords wrapped in angle brackets (< >)                               |
        | **Attributes** | Extra information added to elements (e.g., href, src, alt, id)         |

        | Tag          | Meaning                         |
        | ------------ | ------------------------------- |
        | <!DOCTYPE>   | Tells browser it‚Äôs an HTML5 doc |
        | <html>       | Root element                    |
        | <head>       | Contains metadata, title, links |
        | <title>      | Title shown in browser tab      |
        | <body>       | Visible content of the page     |

    
    üßæ Common HTML Tags
    üîπ Headings
            <h1></h1>
            <h6></h6>
    üîπ Paragraphs and Formatting
            <p>This is a paragraph.</p>
            <strong>Bold</strong>
            <em>Italic</em>
            <br> Line Break
            <hr> Horizontal Line
    üîπ Lists
            <ul>
              <li>Unordered item</li>
            </ul>

            <ol>
              <li>Ordered item</li>
            </ol>
    üîπ Links and Images
            <a href="https://example.com">Click Here</a>
            <img src="image.jpg" alt="Description">
    üîπ Tables
            <table border="1">
              <tr><th>Name</th><th>Age</th></tr>
              <tr><td>Alice</td><td>25</td></tr>
            </table>
    üîπ Forms
            <form>
              <input type="text" placeholder="Enter name">
              <input type="submit" value="Submit">
            </form>

            
    üé® HTML + CSS (for styling)
        HTML gives structure, CSS adds design.
            <p style="color:blue; font-size:18px;">Styled paragraph</p>
        
        
        To link an external CSS:
            <head>
              <link rel="stylesheet" href="styles.css">
            </head>
            
    ‚öôÔ∏è Attributes
        | Attribute | Usage                        |
        | --------- | ---------------------------- |
        | `href`    | Link destination (`<a>`)     |
        | `src`     | Image/video source           |
        | `alt`     | Alternate text for images    |
        | `id`      | Unique identifier            |
        | `class`   | Reusable style group         |
        | `style`   | Inline CSS                   |
        | `type`    | Input type, list type        |
        | `value`   | Default value for inputs     |
        | `name`    | Name used in form submission |
    
    üß© Semantic Tags (for better structure)
        | Tag         | Use Case                         |
        | ----------- | -------------------------------- |
        | <header>    | Top of page or section           |
        | <footer>    | Bottom of page or section        |
        | <nav>       | Navigation bar                   |
        | <section>   | Logical grouping of content      |
        | <article>   | Independent, self-contained unit |
        | <aside>     | Sidebar content                  |
        | <main>      | Main content of the page         |

        Add CSS and JS in HTML
        <!DOCTYPE html>
        <html>
        <head>
          <title>My Page</title>
          <link rel="stylesheet" href="styles.css">  <!-- Link CSS here -->
        </head>
        <body>
          <h1>Click the Button</h1>
          <button onclick="greet()">Click Me</button>
          <script src="script.js"></script>  <!-- Link JavaScript here -->
        </body>
        </html>
```

```python
üé® What is CSS?
    "CSS (Cascading Style Sheets) is a language used to style and design HTML elements."
    "It controls the appearance (colors, fonts, spacing, layout, animations, etc.) of your webpage."
    üí° Why Use CSS?
        HTML = Structure
        CSS = Style
        JavaScript = Behavior

    With CSS, you can:
        Change fonts, colors, and backgrounds
        Control spacing and layout
        Make sites responsive
        Add transitions and animations

    üß± CSS Syntax
        selector {
          property: value;
        }

    üõ†Ô∏è How to Add CSS
    1. Inline CSS
        <p style="color:red;">This is red text</p>
    2. Internal CSS (in <style> tag)
        <head>
          <style>
            p { color: green; }
          </style>
        </head>
    3. External CSS (best practice)
        <link rel="stylesheet" href="styles.css">


        | Property         | Description                     | Example                       |
        | ---------------- | ------------------------------- | ----------------------------- |
        | `color`          | Text color                      | `color: red;`                 |
        | `background`     | Background color/image          | `background: yellow;`         |
        | `font-size`      | Text size                       | `font-size: 20px;`            |
        | `font-family`    | Font type                       | `font-family: Arial;`         |
        | `margin`         | Space outside border            | `margin: 10px;`               |
        | `padding`        | Space inside border             | `padding: 10px;`              |
        | `border`         | Border around element           | `border: 1px solid black;`    |
        | `width`/`height` | Size of element                 | `width: 100px; height: 50px;` |
        | `text-align`     | Horizontal alignment of text    | `text-align: center;`         |
        | `display`        | Layout type (block, flex, etc.) | `display: flex;`              |


        | Selector             | Description                  | Example            |
        | -------------------- | ---------------------------- | ------------------ |
        | `*`                  | Universal selector           | `* { margin: 0; }` |
        | `element`            | Element selector             | `p {}`             |
        | `#id`                | ID selector (unique)         | `#header {}`       |
        | `.class`             | Class selector               | `.button {}`       |
        | `element1, element2` | Multiple elements            | `h1, h2 {}`        |
        | `element child`      | Nested selector (descendant) | `div p {}`         |
        | `element > child`    | Direct child only            | `div > p {}`       |


    üìê Box Model (Important!)
        Each HTML element is treated as a box:
        | margin  |
        | padding |
        | border  |
        | content |

    1. display Property
        | Value    | Use                                   |
        | -------- | ------------------------------------- |
        | `block`  | Default for `<div>`, takes full width |
        | `inline` | Like `<span>`, flows inline           |
        | `flex`   | Flexible layout (modern)              |
        | `grid`   | Grid layout (modern)                  |
        | `none`   | Hides element                         |

    2. Flexbox Example
        .container {
          display: flex;
          justify-content: space-between;
          align-items: center;
        }

    üåç Responsive Design (Media Queries)
        @media screen and (max-width: 600px) {
          body {
            background-color: lightblue;
          }
        }


    üéûÔ∏è CSS Transitions & Animations
    Transition Example
        button {
          transition: background 0.3s ease;
        }

        button:hover {
          background: blue;
        }
    Keyframe Animation
        @keyframes slide {
          from { left: 0; }
          to { left: 100px; }
        }

        .box {
          animation: slide 2s infinite;
        }

    üß© CSS Units
        Unit	Description
        px	Pixels (fixed size)
        %	Percentage of parent
        em	Relative to parent font-size
        rem	Relative to root font-size
        vh/vw	Viewport height/width

    üß† CSS Best Practices
        Always use external CSS for cleaner code
        Use classes, avoid overusing ids
        Keep code organized and commented
        Use shorthand properties when possible (like margin: 10px 20px;)
        Make your designs responsive

    ‚öôÔ∏è Tools & Frameworks
        | Tool/Library        | Use Case                                       |
        | ------------------- | ---------------------------------------------- |
        | **Bootstrap**       | Predefined styles and components               |
        | **Tailwind CSS**    | Utility-first CSS framework                    |
        | **Sass/SCSS**       | CSS preprocessor with variables, nesting, etc. |
        | **Chrome DevTools** | Inspect & edit styles live                     |
        | **Figma**           | UI/UX design and prototyping                   |
```
